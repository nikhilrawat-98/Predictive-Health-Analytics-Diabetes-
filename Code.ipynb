```python
# Import necessary libraries
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split, KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE
import xgboost as xgb
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score, f1_score, roc_curve, precision_recall_curve, auc
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/Users/nick/Downloads/diabetes_prediction_dataset.csv')

# Identify numerical and categorical features
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns.drop('diabetes')
categorical_features = df.select_dtypes(include=['object']).columns

# Data Preprocessing Setup

## Define preprocessing for numerical columns
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

## Define preprocessing for categorical columns
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

## Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Prepare data for model training
X = df.drop('diabetes', axis=1)
y = df['diabetes']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

## Apply preprocessing
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)

# Feature selection using RandomForest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_preprocessed, y_train)
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]
selected_features_indices = indices[:10]  # Select top 10 features
X_train_selected = X_train_preprocessed[:, selected_features_indices]
X_test_selected = X_test_preprocessed[:, selected_features_indices]

# Balancing Dataset with SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_selected, y_train)

# Training and Evaluating XGBoost Model
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train_smote, y_train_smote)
y_pred_xgb = xgb_model.predict(X_test_selected)
y_pred_proba_xgb = xgb_model.predict_proba(X_test_selected)[:, 1]

# Compute false positive rate and true positive rate for ROC curve
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_pred_proba_xgb)

## Calculate metrics
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
roc_auc_xgb = roc_auc_score(y_test, y_pred_proba_xgb)
precision_xgb = precision_score(y_test, y_pred_xgb)
recall_xgb = recall_score(y_test, y_pred_xgb)
f1_xgb = f1_score(y_test, y_pred_xgb)

## Plot ROC Curve for XGBoost
plt.figure(figsize=(8, 6))
plt.plot(fpr_xgb, tpr_xgb, label='XGBoost (area = %0.2f)' % roc_auc_xgb)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for XGBoost')
plt.legend(loc="lower right")
plt.show()

# Defining the DNN Model
def create_model(input_shape):
    model = models.Sequential([
        layers.Dense(128, activation='relu', input_shape=(input_shape,)),
        layers.Dropout(0.2),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Function to plot training/validation accuracy and loss for DNN
def plot_metrics(history, fold_no):
    plt.figure(figsize=(14, 5))
    
    # Plot training & validation accuracy values
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title(f'Model accuracy for fold {fold_no}')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    
    # Plot training & validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title(f'Model loss for fold {fold_no}')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    
    plt.show()

    # Function to calculate and print the metrics for DNN
def print_metrics(y_true, y_pred, y_pred_proba):
    accuracy = accuracy_score(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_pred_proba)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    print(f'Accuracy: {accuracy:.2f}, ROC AUC: {roc_auc:.2f}, Precision: {precision:.2f}, Recall: {recall:.2f}, F1: {f1:.2f}')
    return accuracy, roc_auc, precision, recall, f1

# Function to plot ROC curve for DNN
def plot_roc_curve(y_true, y_pred_proba, fold_no):
    fpr, tpr, _ = roc_curve(y_true, y_pred_proba)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'Fold {fold_no} (area = {roc_auc_score(y_true, y_pred_proba):.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve for DNN Fold {fold_no}')
    plt.legend(loc="lower right")
    plt.show()


# K-Fold Cross-Validation for DNN Evaluation
n_splits = 4
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

fold_no = 1
for train_index, test_index in kf.split(X_train_smote):
    X_train_fold, X_val_fold = X_train_smote[train_index], X_train_smote[test_index]
    y_train_fold, y_val_fold = y_train_smote[train_index], y_train_smote[test_index]

    model = create_model(X_train_fold.shape[1])

    # Training model for the current fold
    print(f'\nTraining on fold {fold_no}...')
    history = model.fit(X_train_fold, y_train_fold, epochs=200, batch_size=64, verbose=2, validation_data=(X_val_fold, y_val_fold))

    # Plotting training and validation metrics
    plot_metrics(history, fold_no)
    
    # Predicting on validation set
    y_pred_proba_val = model.predict(X_val_fold).flatten()

    # Plotting ROC curve for the current fold
    plot_roc_curve(y_val_fold, y_pred_proba_val, fold_no)

    # Printing calculated metrics
    accuracy, roc_auc, precision, recall, f1 = print_metrics(y_val_fold, y_pred_proba_val > 0.5, y_pred_proba_val)
    print(f"Fold {fold_no}: Accuracy = {accuracy:.4f}, ROC AUC = {roc_auc:.4f}, Precision = {precision:.4f}, Recall = {recall:.4f}, F1 Score = {f1:.4f}")

 
    fold_no += 1


# Defining and training the final DNN model with the whole dataset
final_model = create_model(X_train_smote.shape[1])
final_model.fit(X_train_smote, y_train_smote, epochs=200, batch_size=64, verbose=2)

# Save the final model in the TensorFlow  
final_model.save('final_model.keras')

print("Model saved successfully as final_model.keras")
```
